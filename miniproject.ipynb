#loop

```python
data_list = []

for car in results:
    # 1. Name & Year
    name_tag = car.find('a', {"class": "BuyCarCard_carName__SAJVh"})
    car_name = name_tag.get_text(strip=True) if name_tag else "N/A"

    year_match = re.search(r'\((\d{4})', car_name)
    year = year_match.group(1) if year_match else "2025"

    # 2. Price
    price_tag = car.find('p', {"class": "BuyCarCard_priceRange__2q8tm"})
    price = price_tag.get_text(strip=True) if price_tag else "N/A"

    # 3. Fuel & Transmission
    specs = car.find_all('p', {"class": "BuyCarCard_specificationItem__yn5cu"})
    fuel = specs[0].get_text(strip=True) if len(specs) >= 1 else "N/A"
    trans = specs[1].get_text(strip=True) if len(specs) >= 2 else "N/A"

    # 4. Static/Hardcoded fields for New Cars
    km = "0 km"
    owners = "1st Owner"
    loc = "Hyderabad"

    data_list.append({
        "Car Name": car_name,
        "Year": year,
        "Price": price,
        "Fuel": fuel,
        "Transmission": trans,
        "KM Driven": km,
        "Owners": owners,
        "Location": loc
    })

# Create DataFrame
df = pd.DataFrame(data_list)
print(df)
```
